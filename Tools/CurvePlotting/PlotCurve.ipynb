```python
#Curve Plotting

#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
print(x)
```

```python
#defining any arbitrary function

def f(x):
    return (x-4)**2

y = f(x)
```

```python
print(y)
```

```python
#plotting

plt.style.use("seaborn")
plt.plot(x,y)
plt.show()
```
